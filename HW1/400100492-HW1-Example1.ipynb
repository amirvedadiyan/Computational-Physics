## Example 1)
#### Solving equation $\frac{df(x)}{dx} = \cos(x)$ :

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
start = 0
dx = 0.05
stop = 15
# x= start:step:end -> np.arange(start,stop,step)
x = np.arange(start,stop,dx)
n = np.size(x)
```

Define first normal derivative operator as :
$$
\begin{pmatrix}
1 & 0 & 0 & ... & 0\\
-1/\epsilon & 1/\epsilon & 0 & ... & 0\\
0 & -1/\epsilon & 1/\epsilon  & ... & 0\\
...\\
0 & ... & ... & -1/\epsilon & 1/\epsilon 
\end{pmatrix}
$$

```python
D1x = (np.diag(np.ones(n+1),0) - np.diag(np.ones(n),-1))/dx
D1x[0,0] = 1
```

Define $g(x) = \cos(x)$ with initial condition $f_0 = 0$ :

```python
gx = np.cos(1*x[0:])
f0 = 0
# gx = [f0;g] -> np.append(f0,g)
gx = np.append(f0,gx).T
```

Solve equation :

```python
# A*B -> np.matmul(A,B)
ans = np.matmul(np.linalg.inv(D1x),gx)
```

Plot result :

```python
plt.plot(x, gx[1:], label='g(x) = cos(x)')
plt.plot(x, ans[1:], label='f(x) = sin(x)')
plt.plot(x,np.zeros(n), label='y = 0')
plt.title('Plot of g(x) = cos(x) and f(x) = sin(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```
